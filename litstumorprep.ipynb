```python
from pqdm.processes import pqdm
```

```python
images = sorted(glob('/kaggle/input/litsdataset2/images/*.png'))
masks = sorted(glob('/kaggle/input/litsdataset2/masks/*.png'))
```

```python
os.mkdir('tumor-images')
os.mkdir('tumor-masks')
```

```python
def process(image_file):
    mask_file = image_file.replace('images', 'masks')
    image = cv2.imread(image_file, cv2.IMREAD_GRAYSCALE)
    mask = cv2.imread(mask_file, cv2.IMREAD_GRAYSCALE)
    if mask.sum() == 0:
        return None
        
    liver_mask = (mask > 0).astype(np.uint8) * 255
    liver = np.bitwise_and(image, liver_mask)
    
    if np.unique(mask).shape[0] == 2:
        tumor = np.zeros_like(liver)
    else:
        tumor = (mask > 128).astype(np.uint8) * 255
    
    base = os.path.basename(image_file)
    cv2.imwrite(f'tumor-images/{base}', liver)
    cv2.imwrite(f'tumor-masks/{base}', tumor)
```

```python
pqdm(images, process, 8)
```
